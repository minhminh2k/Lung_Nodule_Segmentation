```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = pd.read_csv("../test/image.csv")
mask = pd.read_csv("../test/mask.csv")

print(image)
print(mask)
```

```python
print(image.iloc[:, 1].to_list())
```

```python
bin_number = 40
fig, axs = plt.subplots(2, 2, sharey=True, tight_layout=True, figsize=(10, 10))
axs[0][0].hist(image.iloc[:, 0].to_list(), bins=bin_number)
axs[0][1].hist(image.iloc[:, 1].to_list(), bins=bin_number)
axs[1][0].hist(mask.iloc[:, 0].to_list(), bins=bin_number)
axs[1][1].hist(mask.iloc[:, 1].to_list(), bins=bin_number)
```

```python
fig, ax = plt.subplots(figsize=(10, 10), sharex=True, sharey=True, tight_layout=True)
hist = ax.hist2d(image.iloc[:, 0].to_list(), image.iloc[:, 1].to_list(), bins=10)
```
